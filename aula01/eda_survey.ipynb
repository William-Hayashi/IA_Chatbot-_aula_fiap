# 3. Limpeza de dados
---

- 1. Tratar dados faltantes (remover/preencher)
- 2. Remover dados duplicados

```python
# df = df.dropna()
# df.dropna(inplace=True)
df_clean = df.dropna() #excluir NaN
```
